```python
import pandas as pd
import numpy as np
import os
```

```python
partners = pd.read_csv(path + "/rawdata/imf_import_export_cleaned.csv", dtype={"2019":"float", "2020":"float"})
partners = partners.dropna()
partners = partners[(partners["2019"] > 0) | partners["2020"] > 0]
product = pd.read_csv(path + "/rawdata/merchandise_values_annual_dataset.csv", dtype={"Value":"int", "Year":"category"})
country_code = pd.read_csv(path + "/rawdata/countries_codes_and_coordinates_new.csv")
country_code = country_code.rename({'Country': 'Country_name'}, axis = 1)
```

```python
country_code_r = country_code.merge(partners["from_code"], left_on='Alpha-3code', right_on='from_code', how='right').drop('from_code', axis=1)
country_code_r = country_code_r.drop_duplicates().reset_index(drop=True)
country_code_r
```

```python
country_code_r = country_code.merge(product["ReporterISO3A"], left_on='Alpha-3code', right_on='ReporterISO3A', how='inner').drop('ReporterISO3A', axis=1)
country_code_r = country_code_r.drop_duplicates().reset_index(drop=True)
```

```python
country_code_r = country_code_r.drop_duplicates(subset='Alpha-3code', keep='last')
country_code_r
```

```python
covid_data = pd.read_csv(path + '/rawdata/WHO-COVID-19-global-data.csv')
covid_data
```

```python
covid_data = covid_data.join(country_code_r.set_index('Alpha-2code'), on = 'Country_code', how='right')
covid_data = covid_data[["Country_name", "Date_reported", "Cumulative_cases", "Cumulative_deaths", "Alpha-3code"]]
covid_data = covid_data[covid_data["Date_reported"]=="2020/12/31"].drop('Date_reported', axis=1).reset_index(drop=True)
covid_data
```

```python
whole_df = covid_data.merge(product_total, left_on="Alpha-3code", right_on="ReporterISO3A_", how="left")
whole_df = whole_df[["Alpha-3code", "Country_name", "Cumulative_cases", "Cumulative_deaths", "Import_2020", "Export_2020"]]
whole_df = whole_df.sort_values("Alpha-3code")
whole_df
```

```python
owid_covid['date'] = pd.to_datetime(owid_covid['date'])
owid_covid['year'] = owid_covid['date'].dt.year
owid_covid['quarter'] = owid_covid['date'].dt.to_period('Q')

# owid_df = owid_covid_grouped[owid_covid_grouped['year'] == '2020']

# ### Merge to owid_country and add total cases per population
# owid_df = owid_df.merge(owid_country, how = 'inner', on = 'iso_code')
# owid_df['cases_per_pop_percent'] = (owid_df['total_cases'] / 
#                                     owid_df['population'] * 100)
```

```python
agg = {'total_cases': 'last', 'total_cases_per_million': 'last', 'total_deaths': 'last', 'total_deaths_per_million': 'last',
        'new_cases': 'sum', 'new_cases_per_million' : 'sum', 'new_deaths': 'sum', 'new_deaths_per_million': 'sum',
       'stringency_index': 'mean'}

owid_covid_quarter = (owid_covid.groupby(['iso_code', 'quarter'])
                      .agg(agg).reset_index()).rename(columns={'quarter':'period'})
```

```python
owid_covid_year = (owid_covid.groupby(['iso_code', 'year'])
                      .agg(agg).reset_index()).rename(columns={'year':'period'})
owid_covid_grouped = pd.concat([owid_covid_quarter, owid_covid_year], ignore_index=True, sort=False)
owid_covid_grouped["period"] = owid_covid_grouped["period"].astype(str)
owid_covid_grouped = owid_covid_grouped.sort_values(by=["iso_code", "period"])
```
